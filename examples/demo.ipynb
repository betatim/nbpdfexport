```python
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import pandas as pd
plt.ion()
```

```python
data = np.random.randn(2, 1000)
```

# Using matplotlib for static visualization

```python
fig, ax = plt.subplots()
ax.scatter(*data, c=data[0], s=np.abs(data[1])*50)
```

# Using Altair for interactive visualization

```python
df = pd.DataFrame(data.T, columns=['x', 'y'])
```

```python
alt.Chart(df).mark_point().encode(
    x='x',
    y='y',
    color='x',
    size='y'
)
```
